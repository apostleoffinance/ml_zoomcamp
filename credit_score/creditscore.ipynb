```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as pyplot
%matplotlib inline
```

```python
df = pd.read_csv("/Users/olaoluwatunmise/ML_Zoomcamp/credit_score/CreditScoring.csv")

#turn all columns into a lower case
df.columns = df.columns.str.lower()
df.head()
```

```python
df.status.value_counts()
```

```python
status_values = {
    1: 'ok',
    2: 'default',
    3: 'unk'
}

df.status = df.status.map(status_values)
```

```python
df.marital.value_counts()
```

```python
home_values = {
    1: 'rent',
    2: 'owner',
    3: 'private',
    4: 'ignore',
    5: 'parents',
    6: 'other',
    0: 'unk'
}
df.home = df.home.map(home_values)
 
marital_values = {
    1: 'single', 
    2: 'married', 
    3: 'widow', 
    4: 'separated',
    5: 'divorced',
    0: 'unk'
}
df.marital = df.marital.map(marital_values)
 
records_values = {
    1: 'no',
    2: 'yes',
    0: 'unk'
}
df.records = df.records.map(records_values)
 
job_values = {
    1: 'fixed', 
    2: 'partime', 
    3: 'freelance', 
    4: 'others',
    0: 'unk'
}
df.job = df.job.map(job_values)
 
df.head()
```

### Missing Values

```python
df.describe().round()
```

```python
df.income.max()

 
df.income.replace(to_replace=99999999, value=np.nan)
 
df.income.replace(to_replace=99999999, value=np.nan).max()
```

```python
for c in ['income', 'assets', 'debt']:
    df[c] = df[c].replace(to_replace=99999999, value=np.nan)
 
df.describe().round()
```

```python
df.status.value_counts()

df = df[df.status != 'unk'].reset_index(drop= True)
df
```

### Performing the train/validation/test split

```python
from sklearn.model_selection import train_test_split
 
df_full_train, df_test = train_test_split(df, test_size=0.2, random_state=11)
df_train, df_val = train_test_split(df_full_train, test_size=0.25, random_state=11)
 
df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)
 
df_train.status
```

```python
(df_train.status == 'default').astype('int')
```

```python
y_train = (df_train.status == 'default').astype('int').values
y_val = (df_val.status == 'default').astype('int').values
y_test = (df_test.status == 'default').astype('int').values
 
del df_train['status']
del df_val['status']
del df_test['status']
 
df_train
```

# Decison Trees

```python
def assess_risk(client):
    if client['records'] == 'yes':
        if client['job'] == 'parttime':
            return 'default'
        else:
            return 'ok'
    else:
        if client['assets'] > 6000:
            return 'ok'
        else:
            return 'default'
```

```python
xi = df_train.iloc[0].to_dict()
```

```python
assess_risk(xi)
```

### Train the model

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_extraction import DictVectorizer
from sklearn.metrics import roc_auc_score
```

```python
train_dicts = df_train.fillna(0).to_dict(orient='records')
```

```python
dv = DictVectorizer(sparse=False)
X_train = dv.fit_transform(train_dicts)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
val_dicts = df_val.fillna(0).to_dict(orient = 'records')
X_val = dv.transform(val_dicts)
```

```python
y_pred = dt.predict_proba(X_val)[:, 1]
```

```python
roc_auc_score(y_val, y_pred)
```

```python
y_pred = dt.predict_proba(X_train)[:, 1]
roc_auc_score(y_train, y_pred)
```

```python
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict_proba(X_train)[:, 1]
auc = roc_auc_score(y_train, y_pred)
print('train:', auc)

y_pred = dt.predict_proba(X_val)[:, 1]
auc = roc_auc_score(y_val, y_pred)
print('val:', auc)
```
